```python
from pathlib import Path
import subprocess
import sys

REPO_ROOT = Path("/content/VisScore")
assert (REPO_ROOT / "scripts" / "generate_tufte_synthetic.py").is_file(), (
    f"Missing {REPO_ROOT}/scripts/generate_tufte_synthetic.py — run git clone first or fix REPO_ROOT."
)

subprocess.check_call(
    [sys.executable, "-m", "pip", "install", "-q", "-r", str(REPO_ROOT / "requirements.txt")]
)
```

```python
# REPO_ROOT is defined in the previous cell
REPO_ROOT
```

```python
import os

from huggingface_hub import login

# Prefer Colab secret HF_TOKEN (Runtime → Secrets). Otherwise interactive browser login.
token = None
try:
    from google.colab import userdata

    token = userdata.get("HF_TOKEN")
except Exception:
    pass

if token:
    login(token=token)
    os.environ["HF_TOKEN"] = token
else:
    login()
```

```python
# Optional: save to Google Drive instead of /content
# from google.colab import drive
# drive.mount("/content/drive")
# OUT_DIR = "/content/drive/MyDrive/visscore_synthetic"

OUT_DIR = "/content/visscore_synthetic"
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    str(REPO_ROOT / "scripts" / "generate_tufte_synthetic.py"),
    "--out-dir",
    OUT_DIR,
    "--n-tufte",
    "150",
    "--n-non-tufte",
    "150",
    "--model",
    "flux-schnell",
    "--image-size",
    "768",
    "--seed",
    "42",
]
print(" ".join(cmd))
subprocess.check_call(cmd)
```
